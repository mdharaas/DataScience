```python
#Imputing missing values by mode
df['Material'] = df['Material'].fillna(df['Material'].mode()[0])
df['Transport'] = df['Transport'].fillna(df['Transport'].mode()[0])
df['Remote Location'] = df['Remote Location'].fillna(df['Remote Location'].mode()[0])
```

```python
df['Height'] = df['Height'].interpolate()
df['Weight'] = df['Weight'].interpolate()
df['Width'] = df['Width'].interpolate()
df['Artist Reputation']=df['Artist Reputation'].interpolate()
```

### Regression ML Models

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
```

```python
from sklearn.model_selection import KFold
kf = KFold(n_splits=5,random_state=20,shuffle=False)
kf.split(X) 
```

```python
lR = LinearRegression()
# feeding the training data to the model
LR_accuracy_model = []
for train_index, test_index in kf.split(X):
    # Split train-test
    X_train, X_test = X.iloc[train_index], X.iloc[test_index]
    y_train, y_test = y[train_index], y[test_index]
    # Train the model
    model = lR.fit(X_train, y_train)
    # Append to accuracy_model the accuracy of the model
    LR_accuracy_model.append(r2_score(y_test, model.predict(X_test))*100)

# Print the accuracy    
print(LR_accuracy_model)
```

```python
from sklearn.svm import SVR
# creating the model
svr = SVR()

# feeding the training data to the model
SVR_accuracy_model = []
for train_index, test_index in kf.split(X):
    # Split train-test
    X_train, X_test = X.iloc[train_index], X.iloc[test_index]
    y_train, y_test = y[train_index], y[test_index]
    # Train the model
    model = svr.fit(X_train, y_train)
    # Append to accuracy_model the accuracy of the model
    SVR_accuracy_model.append(r2_score(y_test, model.predict(X_test))*100)

# Print the accuracy    
print(SVR_accuracy_model)
```

```python
from sklearn.ensemble import RandomForestRegressor
rf = RandomForestRegressor(n_estimators = 10, random_state = 10)
R_accuracy_model = []
for train_index, test_index in kf.split(X):
    # Split train-test
    X_train, X_test = X.iloc[train_index], X.iloc[test_index]
    y_train, y_test = y[train_index], y[test_index]
    # Train the model
    model = rf.fit(X_train, y_train)
    # Append to accuracy_model the accuracy of the model
    R_accuracy_model.append(r2_score(y_test, model.predict(X_test))*100)

# Print the accuracy    
print(R_accuracy_model)
```

```python
from xgboost import XGBRegressor
xgb = XGBRegressor()
XGB_accuracy_model = XGB_adjr2score= []

for train_index, test_index in kf.split(X):
    # Split train-test
    X_train, X_test = X.iloc[train_index], X.iloc[test_index]
    y_train, y_test = y[train_index], y[test_index]
    # Train the model
    model = xgb.fit(X_train, y_train)
    # Append to accuracy_model the accuracy of the model
    XGB_accuracy_model.append(r2_score(y_test, model.predict(X_test))*100)
    XGB_adjr2score.append(1-(1-r2_score(y_test,model.predict(X_test)))*((len(X_test)-1)/(len(X_test)-len(X_test.columns)-1)))

# Print the accuracy    
print(XGB_accuracy_model)
#Print the adjusted r2 Score

print("Adj R2 Score: ", XGB_adjr2score)
```
